# Étude en comparaison avec les taux d'intérêt swaption

On étudie ici le rapport $\frac{volatilité ~ observée ~au~ temps~ t}{taux~ d'intérêt ~au~ temps ~t}$

```python
%matplotlib inline
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
from datetime import datetime


table = pd.read_csv('comp_interet.csv', delimiter = ";")
table.fillna(0, inplace=True)
table['Date'] = pd.to_datetime(table['Date'],dayfirst=True)
```

```python
recap = pd.DataFrame()

for i, elem in enumerate(table) :
    if i>0:
        recap = recap.append(table[elem].describe(percentiles = [.05, .25, .5, .75, .95]))
print(recap)
```

Comme constaté dans les autres études, le rapport est de moins en moins volatile lorsque la maturité et le ténor augmentent. 

Les valeurs faibles restent assez proches de la moyenne. En revanche, les grandes valeurs sont très dispersées, on remarque un grand écart entre le quartile à $95\%$ et la valeur maximale.

Le swaption de maturité 1 an et de ténor 5 an a un comportement différent des autres : il est beaucoup plus volatile et en moyenne de valeur supérieure. 

```python
colors = ['RED','GREEN','BLUE','BLACK','ORANGE','PINK','BROWN','YELLOW','CYAN']

fig, ax = plt.subplots()
for i, elem in enumerate(table) :
    if i>0:
        print(str(elem),' : ',colors[i-1])
        ax.plot_date(table['Date'], table[elem],'-',color=colors[i-1])
        fig.autofmt_xdate()
        #plt.show()
```

Ce graphique confirme l'hypothèse formulée ci-dessus : le swaption EUPG15 a un comportement différent des autres. Plus précisement, on voit deux pics sur ce graphique : un premier au premier semestre 2015, un second entre fin 2015 et 2016. Dans la période "normale" avant 2015, le comportement du swaption EUPG15 était similaire à celui des autres. C'est pendant ces deux périodes de comportement ératiques que les comportements sont les plus différents.

L'ordre de grandeur des valeurs prises par EUPG15 est différent de celui des autres swaptions, ce qui nous empêche d'étudier plus précisement leurs comportements ici.

```python
colors = ['RED','GREEN','BLUE','BLACK','ORANGE','PURPLE','BROWN','YELLOW','CYAN']
def graph_mat(table):
    print('Maturité 1 an')
    fig, ax = plt.subplots()
    for i, elem in enumerate(table) :
        if i>0 and i<4:
            print(str(elem),' : ',colors[i-1])
            ax.plot_date(table['Date'], table[elem],'-',color=colors[i-1])
            fig.autofmt_xdate()
    plt.show()
    print('Maturité 5 ans')
    fig, ax = plt.subplots()
    for i, elem in enumerate(table) :
        if i>3 and i<7:
            print(str(elem),' : ',colors[i-1])
            ax.plot_date(table['Date'], table[elem],'-',color=colors[i-1])
            fig.autofmt_xdate()
    plt.show()
    print('Maturité 10 ans')
    fig, ax = plt.subplots()
    for i, elem in enumerate(table) :
        if i>6:
            print(str(elem),' : ',colors[i-1])
            ax.plot_date(table['Date'], table[elem],'-',color=colors[i-1])
            fig.autofmt_xdate()
    plt.show()
    
graph_mat(table)
```

On étudie donc les swaptions séparement en fonction de leur maturité. Les produits de même maturité ont des trajectoires similaires. Plus la maturité est longue, plus le comportement des produits de différents ténors est similaire. La seule divergence notable est la divergence notable est la divergence de EUPG15 lors du choc de fin 2015.

Il est aussi intéressant de noter que le produit le plus volatile est toujours celui de plus faible ténor, sauf pour les swaptions de maturité 10 ans, ou le plus cher est celui de ténor le plus élevé.

Globalement, on identifie 3 périodes : 
1. Une période de comportement "normal", régulier entre mi-2012 et mi-2014
2. Un "pic" entre mi-2014 et mi-2015
3. Une re-normalisation, de valeur moyenne plus élevé jusqu'en 2016

```python
def hist(table):
    for i, elem in enumerate(table) :
        if i>0:
            print(str(elem))
            table[elem].hist()
            plt.show()
hist(table)
```

En observant ces histogrammes, on peut se demander si les distributions ne suivent pas une loi exponentielle

```python
def loi(table,loi='norm'):
    for i, elem in enumerate(table) :
        if i>0:
            test,p=stats.kstest(table[elem], loi)
            if p<0.05:
                print(str(elem)," ne suit pas une loi ",loi)
                
loi(table,'expon')
loi(table)
```

Comme indiqué précedemment, on observe une forte corrélation entre les trajectoires des différents swaptions. On applique donc un test du khi-2 pour voir si les distributions suivent une même loi

```python
ans = pd.DataFrame(index=['EUPG15' 'EUPG110' 'EUPG120' 'EUPG55' 'EUPG510' 'EUPG520' 'EUPG105'
 'EUPG1010' 'EUPG1020'],columns=['EUPG15' 'EUPG110' 'EUPG120' 'EUPG55' 'EUPG510' 'EUPG520' 'EUPG105'
 'EUPG1010' 'EUPG1020'])
def corr(table):
    for i, elem in enumerate(table) :
        for j, ele in enumerate(table) :
            if i>0 and j>0 and i<j:
                test,p=stats.ks_2samp(table[elem], table[ele])
                if p<0.05:
                    pass
                    #print(str(elem),' et ',str(ele),' ne suivent pas la même loi')
                else : 
                    print(str(elem),' et ',str(ele),' suivent la même loi, (p-valeur = ',p,')')
                    fig, ax = plt.subplots()
                    print(str(elem),' : ',colors[1])
                    print(str(ele),' : ',colors[2])
                    ax.plot_date(table['Date'], table[elem],'-',color=colors[1])
                    ax.plot_date(table['Date'], table[ele],'-',color=colors[2])
                    fig.autofmt_xdate()
                    plt.show()

corr(table)
```

Il parait naturel que parmis les swaptions de maturité 5 et 10, ceux de ténors respectifs 10 et 20 suivent la même loi. En effet, on avait vu plus haut que les swaptions de même maturité ont des comportements similaires. De plus, les swaptions aux maturités et ténors plus importants sont moisn volatiles 
, leurs comportements plus "lisses" ont plus de chances d'être similaires.

Plus les maturités/ténors sont élevés, plus la p-valeur est élevée, plus l'hypothèse est "facilement" acceptée.

En revanche, il est plus étonnant de constater que les swaptions de maturité 1 an ténor 5 ans et de maturié 5 an ténor 5 ans suivent la même loi. La p-valeur est d'ailleurs assez faible.

# Étude sur une période de re-normalisation : entre mi-2015 et 2016

Le taux EURIBOR suit une pente descendante relativement constante jusqu'en fin 2015. Il subit alors une courte chute.

```python
tab3 = table[150:]
```

```python
recap = pd.DataFrame()

for i, elem in enumerate(tab3) :
    if i>0:
        recap = recap.append(tab3[elem].describe(percentiles = [.05, .25, .5, .75, .95]))
print(recap)
```

```python
colors = ['RED','GREEN','BLUE','BLACK','ORANGE','PINK','BROWN','YELLOW','CYAN']

fig, ax = plt.subplots()
for i, elem in enumerate(tab3) :
    if i>0:
        print(str(elem),' : ',colors[i-1])
        ax.plot_date(tab3['Date'], tab3[elem],'-',color=colors[i-1])
        fig.autofmt_xdate()
        #plt.show()
```

```python
graph_mat(tab3)
```

Les valeurs sont en moyenne beaucoup plus élevées que précédemment. On observe une "plongée" des valeurs en 2016. Ce choc est particulièrement sensible pour le swaption de maturité 1 an et ténor 5 ans. Sa variance est très important sur cette période alors que les autres ont des comportements plus réguliers.

```python
corr(tab3)
```

```python
hist(tab3)
```

On ne fera pas de test d'adéquation ici, visiblement, aucun swaption ne suit de loi normale ou exponentielle

On peut se demander si après la période de pic de mi-2014 à mi-2015, les swaptions et taux d'intérêt retrouvent leur comportement d'avant crise. On a vu que les valeurs étaient en moyenne plus élevées dans la troisième période, on normalise donc les valeurs puis on tente de voir si les distributions suivent une même loi :

```python
tab1_B = tab1
for i, elem in enumerate(tab1):
    if i>0:
        m=np.mean(tab1[elem])
        tab1_B[elem]=tab1[elem]-m
        
tab3_B = tab3
for i, elem in enumerate(tab3):
    if i>0:
        m=np.mean(tab3[elem])
        tab3_B[elem]=tab3[elem]-m

for i, elem in enumerate(tab1_B) :
    if i>0 :
        test,p=stats.ks_2samp(tab1_B[elem], tab3_B[elem])
        if p<0.05:
            print("Les comportements sur les deux périodes ne sont pas similaires pour ",str(elem))
            print("p-valeur du test : ",p)
```
