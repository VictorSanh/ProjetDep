```python
%matplotlib inline 
import numpy as np # module pour les outils math√©matiques
import matplotlib.pyplot as plt # module pour les outils graphiques
from scipy import interpolate
from IPython.display import Image
```

```python
x = np.arange(0, 2*np.pi+np.pi/4, 2*np.pi/8)
y = np.sin(x)
tck = interpolate.splrep(x, y, s=0.2)
xnew = np.arange(0, 2*np.pi, np.pi/50)
ynew = interpolate.splev(xnew, tck, der=0)
```

```python
plt.figure()
plt.plot(x, y, 'x', xnew, ynew, xnew, np.sin(xnew), x, y, 'b')
plt.legend(['Linear', 'Cubic Spline', 'True'])
plt.axis([-0.05, 6.33, -1.05, 1.05])
plt.title('Cubic-spline interpolation')
plt.show()
```

```python
x = np.arange(0, 7, 1)
y = 3*x**3 + 2*x**2 - 3*x + 10
interpolante = interpolate.interp1d(x, y, kind = 4) #On interpole avec un polynome de degr√© kind, ici kind vaut 4.
xnew = np.arange(0, 7, 1)
ynew = interpolante(xnew)
```

```python
plt.figure()
plt.plot(x, y, 'o', xnew, ynew, '-')
plt.show()
```

```python
epsilon = 1e-5 #C'est suffisament comme pas.

#D√©riv√©e par diff√©rence finie
def evalDerive(interpolante, x):
    return (interpolante(x+epsilon) - interpolante(x-epsilon))/(2*epsilon)

def evalDeriveSeconde(interpolante, x):
    return (interpolante(x+epsilon) + interpolante(x-epsilon) - 2*interpolante(x))/(epsilon*epsilon)

evalDerive(interpolante, 3), evalDeriveSeconde(interpolante, 3)
```
